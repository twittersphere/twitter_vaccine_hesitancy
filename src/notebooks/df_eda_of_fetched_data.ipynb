```python
import pandas as pd

from src.scripts.read_data import ReadData
```

```python
from nltk.tokenize import RegexpTokenizer
TOKENIZER = RegexpTokenizer(r'\w+')
```

```python
world_data_path = "/data/raw/daily_world_en_csv"
us_data_path = "/data/raw/US_data_csv_joined_with_locations"
dataframes_path = "/data/processed/data_frames"
```

```python
read_data_world = ReadData(world_data_path, ['id', 'text', 'author_id'], filter_tweets=True)
read_data_world.read_csvs_and_combine_data()
```

```python
read_data_us = ReadData(us_data_path, ['id', 'text', 'latitude'], filter_tweets=True)
read_data_us.read_csvs_and_combine_data()
```

```python
num_of_tweets = read_data_world.data.shape[0]
num_of_author_ids = read_data_world.data['author_id'].unique().shape[0]
num_of_locational_tweets = read_data_us.data.shape[0]

numbers = [num_of_tweets, num_of_author_ids, num_of_locational_tweets]
labels = ['Tweets', 'Users', 'Tweets with Geo-Location']

df_duplicates_removed = pd.DataFrame({'Counts': numbers, 'Labels': labels})
df_duplicates_removed
```

```python
df_duplicates_removed.to_parquet(f"{dataframes_path}/eda_df.parquet", index=False)
```
