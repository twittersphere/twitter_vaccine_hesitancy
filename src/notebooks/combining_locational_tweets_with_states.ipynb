```python
import os
import pandas as pd
from tqdm import tqdm
```

```python
us_tweets_path = "/data/raw/daily_US_en_csv"
geo_locations_path = "/data/raw/unique_geo_ids_with_states.csv"
saving_path = "/data/raw/US_data_csv_joined_with_locations"
```

```python
if not os.path.exists(saving_path):
    os.mkdir(saving_path)
```

```python
geo_locations = pd.read_csv(geo_locations_path)
```

```python
raw_tweets_name_list = os.listdir(us_tweets_path)
raw_tweets_name_list.sort()
```

```python
for file_name in tqdm(raw_tweets_name_list):
    df = pd.read_csv(f"{us_tweets_path}/{file_name}")
    df = df.drop(columns=['longitude', 'latitude', 'coordinates_type'])
    joined_df = df.join(geo_locations.rename(columns={'id':'geo_place_id'}).set_index('geo_place_id'), on='geo_place_id')
    joined_df.to_csv(f"{saving_path}/{file_name}", index=False)
```
