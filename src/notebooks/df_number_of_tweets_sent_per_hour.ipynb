```python
import pandas as pd
from datetime import timedelta

from src.scripts.read_data import ReadData
```

```python
dataframes_path = "/data/processed/dataframes"
```

```python
world_data_path = "/data/raw/daily_world_en_csv"

read_data_world = ReadData(world_data_path, ['id', 'created_at', 'text'], filter_tweets=True)
read_data_world.read_csvs_and_combine_data()
```

```python
read_data_world.data = read_data_world.data.drop(columns=['text'])
read_data_world.data['created_at'] = pd.to_datetime(read_data_world.data['created_at']) - timedelta(hours=5)
```

```python
df = read_data_world.data.copy(deep=True)
df = df.set_index('created_at')
```

```python
hours = [f'{i:02d}:00' for i in range(24)] + ['00:00']
tweet_counts = [df.between_time(hours[idx], hours[idx+1]).shape[0] for idx in range(len(hours)-1)]

plot_df = pd.DataFrame({'hour': hours[:-1], 'tweet_count':tweet_counts})
```

```python
plot_df.to_parquet(f'{dataframes_path}/time_tweet_number_analysis.parquet')
```
