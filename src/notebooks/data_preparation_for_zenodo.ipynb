```python
import os
import pandas as pd
```

```python
num_threads = os.cpu_count()

raw_data = "/data/raw"
probs_path = "/data/processed/CTM/probs"
dataframes_path = "/data/processed/data_frames"
```

### Attitude Data

```python
world_data_sentiments_raw = pd.read_parquet(f"{dataframes_path}/world_data_sentiments_raw.parquet")
world_data_sentiments_raw = world_data_sentiments_raw.drop(columns=['created_at'])
world_data_sentiments_raw.to_parquet(f"{dataframes_path}/zenodo/attitude_predictions.parquet", index=False)
```

### CTM Data

```python
ctm_probs = pd.read_parquet(f"{probs_path}/cleaned_probs_8_200_0.2.parquet")
cleaned_world_anti_tweets = pd.read_parquet(f"{raw_data}/cleaned_world_anti_tweets_and_ids.parquet")
```

```python
ctm_probs['id'] = cleaned_world_anti_tweets['id'].values
ctm_probs[['id', *ctm_probs.columns[:-1]]].to_parquet(f"{dataframes_path}/zenodo/ctm_predictions.parquet", index=False)
```
