```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataframes_path = "data/processed/data_frames"
supp_figures_path = "figures/supp_figure_svgs"
```

```python
plot_df = pd.read_parquet(f'{dataframes_path}/time_tweet_number_analysis.parquet')
```

```python
sns.set_context('poster')
sns.set_style("whitegrid")

fig, ax = plt.subplots(figsize=(16, 8))
sns.barplot(data=plot_df, x='hour', y='tweet_count', ax=ax, color='gray')
ax.set_xlabel('EST Hour')
ax.set_ylabel('Tweet Count')
plt.xticks(rotation=90, ha='right')
fig.tight_layout()
fig.savefig(f"{supp_figures_path}/figure1.svg")
```
