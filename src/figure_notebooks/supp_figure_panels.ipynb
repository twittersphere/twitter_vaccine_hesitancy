```python
pw.clear()
ax1 = pw.Brick("ax11", figsize=(7, 7))
ax1.imshow(plt.imread(f"{supp_figures_path}/timebased_atv_lineplot_locational_nonlocational.png"))
ax1.axis('off')
ax1.set_title('(a)', fontsize=24)
ax1.title.set_x(0)

ax2 = pw.Brick("ax12", figsize=(7, 7))
ax2.imshow(plt.imread(f"{supp_figures_path}/correlation_of_locational_and_nonlocational_tweets.png"))
ax2.axis('off')
ax2.set_title('(b)', fontsize=24)
ax2.title.set_x(0)

ax = ax1 / ax2
ax.savefig()

ax.savefig(f"{panel_figures_path}/supp_figure3.pdf")
```
