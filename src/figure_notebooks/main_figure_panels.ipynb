```python
import patchworklib as pw
import matplotlib.pyplot as plt
```

```python
main_figures_path = "/figures/experimental/main_figures"
panel_figures_path = "/figures/paper_ready/main_figures"
```

### Figure 1

```python
pw.clear()
ax1 = pw.Brick("ax11", figsize=(10, 10))
ax1.imshow(plt.imread(f"{main_figures_path}/tweet_user_geoloc_counts.png"))
ax1.axis('off')
ax1.set_title('(a)', fontsize=40, loc='left')
ax1.title.set_x(0)

ax2 = pw.Brick("ax12", figsize=(10, 10))
ax2.imshow(plt.imread(f"{main_figures_path}/precision_recall_curve_threshold_99.png"))
ax2.axis('off')
ax2.set_title('(b)', fontsize=40)
ax2.title.set_x(0)

ax3 = pw.Brick("ax21", figsize=(10, 10))
ax3.imshow(plt.imread(f"{main_figures_path}/atv_stachgraph.png"))
ax3.axis('off')
ax3.set_title('(c)', fontsize=40)
ax3.title.set_x(0)

ax = (ax1 | ax2) / ax3
ax.savefig()

ax.savefig(f"{panel_figures_path}/paper_figure1.pdf")
```
